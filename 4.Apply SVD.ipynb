## Apply SVD over the picture of your face

```python
from functions.svd import SVD
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

### Load my picture

```python
img = mpimg.imread('myPhotoGray.png')
```

###  Progressively increasing the number of singular values used

```python
vectors = [1, 5, 14, 30, 100, 200]

plt.figure(figsize=[10,7])

for i in range(6):
    plt.subplot(2,3,i+1)
    svd = SVD(n_vectors=vectors[i])
    new_image=svd.fit_transform(img)
    reverse=svd.inverse_transform()
    plt.imshow(reverse,cmap='gray')
    
    plt.title("Components: "+str(vectors[i]))

plt.subplots_adjust(wspace=0.2, hspace=0.2)
plt.show()
```
