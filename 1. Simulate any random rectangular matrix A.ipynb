```python
import numpy as np
from numpy.linalg import eigh
```

## 1. Simulate any random rectangular matrix A.


```python
A= np.asmatrix(np.random.rand(3, 4))
A
```

### 1.1
Rank of matrix A: The rank of a rectangular matrix is the maximum number of linearly independent rows or columns of the matrix. To find the rank of a rectangular matrix using Python, we can use the numpy.linalg.matrix_rank()

```python
rank = np.linalg.matrix_rank(A)
print('Then rank of the matrix with shape {} is {}'.format(A.shape,rank))
```

### 1.2
The Determinant of a rectangular matrix does not exist, that is why when using np.linalg.det it gives an error

```python
import traceback
try:
    np.linalg.det(A)
except:
    traceback.print_exc(limit=1)
```

###  1.3
Invert a matrix using numpy.linalg.  Invert A: Using np.linalg Compute the (Moore-Penrose) pseudo-inverse of a matrix. 

```python
A_inv = np.linalg.pinv(A)
A_inv
```

###  1.4
How are eigenvalues and eigenvectors of A’A and AA’ related?  

Eigenvalues--> For this matrix I will use np.linalg.eig.  eig works for general matrices and therefore uses a slower algorithm

```python
print(np.linalg.eig(np.dot(A.T, A))[0])
print(np.linalg.eig(np.dot(A, A.T))[0])
```

### What interesting differences can you notice between both?

The value are different because Rectangular matrices do not have eigenvectors/eigvalues. Instead they have singular values & singular vectors. (Square matrices have both - eigenvalues and singular values).
